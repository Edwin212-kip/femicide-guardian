## Data Preprocessing and Validation

```python
# Initialize NLTK components
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stop words and lemmatize
    processed_tokens = [
        lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words
    ]
    # Join tokens back into a string
    return " ".join(processed_tokens)

# Apply the preprocessing function to the 'text' column
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
# Convert the processed text column into TF-IDF features
tfidf = TfidfVectorizer(max_features=5000)
X_text = tfidf.fit_transform(df['processed_text'])
```

```python
# Select your categorical and numeric features from the dataframe
categorical_features = ['Location', 'suspect relationship', 'Type of femicide', 
                        'Murder Scene', 'Mode of killing', 'Circumstance']
numeric_features = ['Age']

# Create a new DataFrame containing only those features
X = df[numeric_features + categorical_features].copy()
```

```python
# Encode the categorical features into numbers
categorical_features = ['Location', 'suspect relationship', 'Type of femicide', 
                        'Murder Scene', 'Mode of killing', 'Circumstance']
for col in categorical_features:
    encoder = LabelEncoder()
    X[col] = encoder.fit_transform(X[col])
```

```python
# Encode the target variable and store the encoder
le = LabelEncoder()
y = le.fit_transform(df['Type of femicide']) 
```

```python
X = df.drop(columns=['Verdict', 'text', 'published_date', 'country_name', 'type of murder'])
y = df['Type of femicide']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_text, y, test_size=0.3, random_state=42, stratify=y)
```

## Modeling

### Logistic regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Define grid
log_params = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

log_clf = LogisticRegression(random_state=42)

log_grid = GridSearchCV(log_clf, log_params, cv=5, scoring='recall', n_jobs=-1, verbose=1)
log_grid.fit(X_train, y_train)

# Best model
best_log = log_grid.best_estimator_
print("Best Logistic Regression Parameters:", log_grid.best_params_)
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}

rf_clf = RandomForestClassifier(random_state=42)

rf_grid = GridSearchCV(rf_clf, rf_params, cv=5, scoring='recall', n_jobs=-1, verbose=1)
rf_grid.fit(X_train, y_train)

best_rf = rf_grid.best_estimator_
print("Best Random Forest Parameters:", rf_grid.best_params_)
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'min_samples_split': [2, 5]
}

gb_clf = GradientBoostingClassifier(random_state=42)

gb_grid = GridSearchCV(gb_clf, gb_params, cv=5, scoring='recall', n_jobs=-1, verbose=1)
gb_grid.fit(X_train, y_train)

best_gb = gb_grid.best_estimator_
print("Best Gradient Boosting Parameters:", gb_grid.best_params_)
```

## Evaluation


```python
def evaluate_model(model, name, target_names=None):
    y_pred = model.predict(X_test)
    
    print(f"\n--- {name} Evaluation ---")
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))


```

```python
evaluate_model(best_log, "Logistic Regression", target_names=le.classes_)
evaluate_model(best_rf, "Random Forest", target_names=le.classes_)
evaluate_model(best_gb, "Gradient Boosting", target_names=le.classes_)
```
