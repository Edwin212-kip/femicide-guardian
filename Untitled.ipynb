Despite frequent media reports, data science has not been widely used to uncover trends,
identify hotspots, or support prevention efforts

# Goal 
To analyze and predict femicide patterns in Kenya using available data.

- To conduct advanced exploratory data analysis (EDA), including temporal trend detection, geospatial mapping, and demographic profiling of both victims and perpetrators.

```python
### Preprocessing


import re

def clean_text(text):
    if pd.isnull(text):
        return ""
    # Lowercase, remove punctuation, numbers, extra spaces
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df.loc[:, 'Circumstance_clean'] = df['Circumstance'].apply(clean_text)
df.loc[:, 'text_clean'] = df['text'].apply(clean_text)

```

```python
### Preprocessing


import re

def clean_text(text):
    if pd.isnull(text):
        return ""
    # Lowercase, remove punctuation, numbers, extra spaces
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df.loc[:, 'Circumstance_clean'] = df['Circumstance'].apply(clean_text)
df.loc[:, 'text_clean'] = df['text'].apply(clean_text)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
X_circumstance = tfidf.fit_transform(df['Circumstance_clean'])
X_text = tfidf.fit_transform(df['text_clean'])
```

```python

from textblob import TextBlob

df.loc[:, 'Circumstance_sentiment'] = df['Circumstance_clean'].apply(lambda x: TextBlob(x).sentiment.polarity)
df.loc[:, 'text_sentiment'] = df['text_clean'].apply(lambda x: TextBlob(x).sentiment.polarity)


```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english', max_features=1000)
X_topics = vectorizer.fit_transform(df['Circumstance_clean'])

lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda_topics = lda.fit_transform(X_topics)

# Add the top topic as a new feature
df.loc[:,'top_topic'] = lda_topics.argmax(axis=1)
```

```python
def dig_femicide_insights(df, target_col=None):
    print(" Dataset Shape:", df.shape)
    print("\n Column Types:\n", df.dtypes)
    
    print("\n Missing Values:\n", df.isnull().sum())

    print("\n First 5 Rows:\n", df.head())
    
    print("\n Value Counts (Top Categorical Columns):")
    for col in df.columns:
        if df[col].dtype == 'object' or df[col].nunique() < 30:
            print(f"\n🔸 {col}:\n{df[col].value_counts(dropna=False)}")

    print("\n Numerical Description:\n", df.describe())
    
    if target_col:
        print(f"\n Target Class Balance ({target_col}):\n", df[target_col].value_counts(normalize=True))
    
    print("\n Duplicate Rows:", df.duplicated().sum())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_femicide_data(df):
    print(" Basic Info:")
    print(df.info())
    print("\n Missing Values:")
    print(df.isna().sum())

    print("\n Duplicate Rows:", df.duplicated().sum())

    # ── Distribution of Age
    plt.figure(figsize=(10, 4))
    sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='orange')
    plt.title("Age Distribution of Victims")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

    # ── Circumstance Sentiment Distribution
    plt.figure(figsize=(10, 4))
    sns.countplot(x='Circumstance_sentiment', data=df, order=sorted(df['Circumstance_sentiment'].dropna().unique()))
    plt.title("Circumstance Sentiment Levels")
    plt.show()

    # ── Text Sentiment Distribution
    plt.figure(figsize=(10, 4))
    sns.histplot(df['text_sentiment'].dropna(), bins=20, kde=True)
    plt.title("Distribution of Text Sentiment from Stories")
    plt.xlabel("Text Sentiment Score")
    plt.show()

    # ── Top Topic Frequency
    plt.figure(figsize=(10, 4))
    sns.countplot(x='top_topic', data=df)
    plt.title("Most Common Topics Raised")
    plt.xlabel("Topic Code")
    plt.ylabel("Frequency")
    plt.show()

    # ── Heatmap for Age vs Sentiment
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Age', y='Circumstance_sentiment', hue='top_topic', data=df, palette='cool')
    plt.title("Age vs Circumstance Sentiment Colored by Topic")
    plt.show()

    print("\n✅ Analysis complete. Ready to label & train.")
```

```python
analyze_femicide_data(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def insight_miner(df):
    print("FEMICIDE DATA INSIGHT REPORT")
    print("==================================")
    # --- SENTIMENT ANALYSIS ---
    print("\n🧠 Circumstance Sentiment Breakdown:")
    print(df['Circumstance_sentiment'].value_counts().sort_index())

    print("\n🧠 Text Sentiment Breakdown:")
    print(df['text_sentiment'].value_counts().sort_index())

    # --- NUMERICAL STATS ---
    print("\n📐 NUMERICAL DESCRIPTION:")
    print(df[['Age', 'Circumstance_sentiment', 'text_sentiment']].describe())

    # --- CROSS-TABS ---
    print("\n📊 RELATIONSHIP BETWEEN SENTIMENT & SUSPECT RELATIONSHIP:")
    print(pd.crosstab(df['suspect relationship'], df['Circumstance_sentiment']))

    print("\n📊 TOPIC vs. MODE OF KILLING:")
    print(pd.crosstab(df['top_topic'], df['Mode of killing']))

    # --- OUTLIERS ---
    print("\n🚨 EXTREME CASES (Young victims under 10 or age above 80):")
    print(df[(df['Age'] < 10) | (df['Age'] > 80)][['Age', 'Location', 'Mode of killing', 'Circumstance']])

    print("\n🚨 EXTREME SENTIMENT CASES:")
    extreme_sent = df[(df['Circumstance_sentiment'].abs() > 0.5) | (df['text_sentiment'].abs() > 0.5)]
    print(extreme_sent[['Circumstance_sentiment', 'text_sentiment', 'Circumstance', 'text']])

    # --- VISUALS (Optional) ---
    sns.histplot(df['Age'].dropna(), bins=20, kde=True)
    plt.title("📊 Age Distribution of Victims")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()

    sns.boxplot(x='suspect relationship', y='Circumstance_sentiment', data=df)
    plt.title("📊 Sentiment vs Suspect Relationship")
    plt.xticks(rotation=45)
    plt.show()

    print("\n✅ Insight Miner Report Complete.")

```
